```python
# import library
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# load dataset
data = pd.read_csv('cancer.csv')
data.head()
```

```python
# check data type
data.info()
```

```python
# delete column id & unnamed
data.drop(columns=['id','Unnamed: 32'], axis=1, inplace=True)
data.columns
```

```python
# check new data 
data.info()
```

```python
# encoding diagnosis column with label encoding
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

data['diagnosis'] = le.fit_transform(data['diagnosis'])
```

```python
# quick look data
data.head()
```

```python
# diagnosis after label encoding
data.diagnosis.unique()
```

```python
# split data feature * target
from sklearn.model_selection import train_test_split


x = data.drop(['diagnosis'],axis=1)
y = data['diagnosis']

x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=42)
```

```python
# modeling
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(criterion='gini',random_state=42)

model.fit(x_train, y_train)
y_pred = model.predict(x_test)
```

```python
# compare actual * prediction
compare = pd.DataFrame({'Actual' : y_test, 'Prediction' : y_pred})
compare
```

```python
# metrics
from sklearn.metrics import confusion_matrix, classification_report
```

```python
confusion_matrix(y_test,y_pred)
```

```python
sns.heatmap(confusion_matrix(y_test,y_pred),annot=True)
plt.ylabel('Actual')
plt.xlabel('Prediction')
```

```python
print(classification_report(y_test,y_pred))
```
